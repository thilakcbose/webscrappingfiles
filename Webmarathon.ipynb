```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import openpyxl

ranks =[]
brands = []
prices = []
ratings = []
counts= []
posters = []

for i in range(1,3):
    content = requests.get("https://www.amazon.in/gp/bestsellers/watches/ref=zg_bs_pg_1_watches?ie=UTF8&pg="+str(i))
    soup = BeautifulSoup(content.text,'html.parser')
    watches = soup.find_all('div', class_ = "_cDEzb_iveVideoWrapper_JJ34T")

    for watch in watches:
        try:
            rank = watch.find('span',class_= "zg-bdg-text").text
            
            brand = watch.find('div', class_="_cDEzb_p13n-sc-css-line-clamp-3_g3dy1").text.split()[0]
        
            # price = watch.find('span', class_="_cDEzb_p13n-sc-price_3mJ9Z").text
            rating = float(watch.find('span',class_="a-icon-alt").text.split()[0])
            count = int(watch.find('span',class_="a-size-small").text.replace(',',''))
            poster=watch.find('img',class_="a-dynamic-image p13n-sc-dynamic-image p13n-product-image")['src']
        except AttributeError:
                pass
        print(rank,brand,rating,count,poster)

        ranks.append(rank)
        brands.append(brand)
        ratings.append(rating)
        counts.append(count)
        posters.append(poster)

    # creating a dictionary
    data = {"RANK": ranks,
            "BRAND": brands,
            "RATING":ratings,
            "REVIEW_COUNT":counts,
            "POSTER_LINK":posters}

    # creating a dataframe using dict
    df = pd.DataFrame(data)
    print(df.tail())

    # Save the DataFrame to an Excel file (XLSX format)
    df.to_excel('Watches.xlsx', index=False)

 


```

```python
import openpyxl
wb = openpyxl.load_workbook("Watches.xlsx")
ws = wb.active

# Add the HYPERLINK formula to the existing poster link values in the same column
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=5, max_col=5):
    for cell in row:
        cell.value = f'=HYPERLINK("{cell.value}", "{cell.value}")'

wb.save("Watches.xlsx")
```
