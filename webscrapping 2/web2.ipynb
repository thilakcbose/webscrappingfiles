```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Initialize empty lists for each column
ranks = []
names = []
years = []
ratings = []
review_counts = []
poster_links = []

headers = {
    'User-Agent': 'Your User Agent Here'
}

content = requests.get("https://www.imdb.com/chart/top/", headers=headers)
soup = BeautifulSoup(content.text, 'html.parser')
movies = soup.find_all('li',class_="ipc-metadata-list-summary-item sc-1364e729-0 caNpAE cli-parent")

for movie in movies:
    rank = int(movie.find('h3', class_="ipc-title__text").text.split('.')[0])
    name = movie.find('h3', class_="ipc-title__text").text.split('.')[-1]
    year = int(movie.find('span', class_="sc-be6f1408-8 fcCUPU cli-title-metadata-item").text)
    rating = movie.find('span', class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating").text
    rating1 = float(rating.split('\xa0')[0])
    count = (movie.find('span', class_="ipc-rating-star--voteCount").text.replace('(', '').replace(')', ''))
    link = movie.find('img',class_="ipc-image")
    poster = link['src']

    print(rank,name,year,rating1,count,poster)
    
    ranks.append(rank)
    names.append(name)
    years.append(year)
    ratings.append(rating)
    review_counts.append(count)
    poster_links.append(poster)


# Create a DataFrame using the lists
df = pd.DataFrame({
    'RANK': ranks,
    'NAME': names,
    'YEAR': years,
    'RATING': ratings,
    'REVIEW_COUNT': review_counts,
    'POSTER_LINK': poster_links
})

# Save the DataFrame to an Excel file (XLSX format)
df.to_excel('movies_dataset.xlsx', index=False)



```

```python
import openpyxl
wb = openpyxl.load_workbook("movies_dataset.xlsx")
ws = wb.active

# Add the HYPERLINK formula to the existing poster link values in the same column
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=6, max_col=6):
    for cell in row:
        cell.value = f'=HYPERLINK("{cell.value}", "{cell.value}")'

wb.save("movies_dataset.xlsx")
```
